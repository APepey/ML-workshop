# Joining datasets

```python
demographics = pd.read_csv("../data/workshop/demographics.csv")
demographics.head()
```

```python
deaths = pd.read_csv("../data/workshop/malaria-deaths.csv")
deaths.head()
```

```python
incidence = pd.read_csv("../data/workshop/incidence-of-malaria.csv")
incidence.head()
```

```python
print(incidence.shape, deaths.shape, demographics.shape)
```

```python

predf = pd.merge(incidence, deaths, on="country", how='right')
predf.head(2)
```

```python
df = pd.merge(predf, demographics, on="country", how='right')
df.head(2)
```

```python
df.to_csv("../data/workshop/worldbank-df.csv")
```
